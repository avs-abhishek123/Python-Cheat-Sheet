Dictionary in Python is an ordered collection of data values, used to store data values like a map, which, unlike other Data Types that hold only a single value as an element, Dictionary holds key:value pair. Key-value is provided in the dictionary to make it more optimized. 

1. Dictionaries are unordered & mutable
2. Keys are unique ( No duplicate key is allowed )
3. Keys must be immutable
4. Dictionary comprehension
5. Dictionary Keys are Case Sensitive


* A dictionary contains key-value pairs but does not possess an order for the pairs. 

* Thus, a more precise definition is that a dictionary is an unordered collection of key-value pairs.

* As a result, we cannot perform typical operations associated with the order of items. 

#  **For instance, it is not possible to get the first or last item in a dictionary.**

```python
mydict={"name": "Max", "age": 28, "city": "Newyork"}
print(mydict)

mydict2=dict(name="Mary", age=27, city="Boston")
print(mydict2)
```

Delete elements

```python
# Removing elements from a dictionary

# create a dictionary
squares = {1: 1, 2: 4, 3: 9, 4: 16, 5: 25}

# remove a particular item, returns its value
# Output: 16
print(squares.pop(4))

# Output: {1: 1, 2: 4, 3: 9, 5: 25}
print(squares)

# remove an arbitrary item, return (key,value)
# Output: (5, 25)
print(squares.popitem())

# Output: {1: 1, 2: 4, 3: 9}
print(squares)

# remove all items
squares.clear()

# Output: {}
print(squares)

# delete the dictionary itself
del squares

# Throws Error
#print(squares)
```

.copy()

```python
mydict={"name": "Max", "age": 28, "city": "Newyork"}
mydict_cpy=mydict.copy()
print(mydict)
print(mydict_cpy)
print()
mydict_cpy["email"]="max@xyz.com"
print(mydict_cpy)
print(mydict)#original list doesn't get modified
```

dict()

```python
mydict={"name": "Max", "age": 28, "city": "Newyork"}
mydict_cpy=dict(mydict)
print(mydict)
print(mydict_cpy)
print()
mydict_cpy["email"]="max@xyz.com"
print(mydict_cpy)
print(mydict)#original list doesn't get modified
```
